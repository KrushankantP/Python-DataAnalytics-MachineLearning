## Q2. 
Group the data by **Team** and **Year**, display each team’s, each year’s **TeamPoints** and **OpponentPoints.**

```python
from datetime import datetime
nba['Date']=pd.to_datetime(nba['Date'])
gby_team= nba.groupby([nba['Team'],nba['Date'].dt.year])[['TeamPoints' ,'OpponentPoints']].mean()
gby_team
```

## Q3.
Create another DataFrame named nba_2017 where it contains records of year 2017. In this DataFrame, you should have 2,536 rows.

```python
nba_2017 =nba[nba['Date'].dt.year == 2017]
```

```python
nba_2017.head()
```

## Q4. 
Draw a bar chart for **all teams’ average** **TeamPoints** per game in 2017. Add an appropriate x_label, y_label and title to the chart (NO SUCH THINGS IN THE FOLLOWING EXAMPLE FIGURE). Use an appropriate size for the chart.

```python
bar_data = nba_2017.groupby(nba['Team'])['TeamPoints'].mean()
```

```python
ax=bar_data.plot.bar(figsize=(12,8))
ax.set_xlabel('Team in 2017')
ax.set_ylabel('Average Team')
ax.set_title('Bar chart of Average Team Point in 2017', size=20)
```

## Q5.
Display the **top 5 teams** in **2017** with the best defending in terms of **least OpponentPoints**. The output should contain the Team’s name and the average OpponentPoints per game.

```python
top5team=nba_2017.groupby('Team')['OpponentPoints'].mean()
top5team.sort_values().head()
```

## Q6. 
Draw a stacked column chart showing the numbers of **W and L** for **each team**. Add an appropriate x_label, y_label and title to the chart (NO SUCH THINGS IN THE FOLLOWING EXAMPLE FIGURE). Use an appropriate size for the chart.

```python
nowl=pd.crosstab(nba_2017['Team'], nba_2017['WINorLOSS'])
nowl
```

```python
ax=nowl.plot.barh(stacked=True, figsize=(12,8))
ax.set_xlabel('Win and loass Points')
ax.set_ylabel('Teams')
ax.set_title('Bar chart Of W & L Point in 2017', size=20)
```

## Q7.
There is an important metric to evaluate a **team’s performance** which is the **difference** between **it scores** and **its opponent scores.** Show all teams with this metric from the largest to the smallest.

```python
team_mean =nba_2017.groupby('Team').mean()
diff_score =team_mean['TeamPoints']-team_mean['OpponentPoints']
diff_score.sort_values(ascending =False)
```
