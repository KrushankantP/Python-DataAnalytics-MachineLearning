```python
fig, ax= plt.subplots(2,2)
ax[0,0].plot(np.random.randn(50).cumsum(),'r--')
ax[0,1].hist(np.random.randn(100), bins=20, color='k')
ax[1,0].scatter(np.random.randn(30), np.random.randn(30))
```

```python
fig, ax= plt.subplots(2,2)
plt.subplots_adjust(wspace=0,hspace=0)
ax[0,0].plot(np.random.randn(50).cumsum(),'r--')
ax[0,1].hist(np.random.randn(100), bins=15, color='k')
ax[1,0].scatter(np.random.randn(30), np.random.randn(30))
```

```python
#in this example, we put multiple lines into one chart
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label ='default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

```python
#in this exapmple we will see how to set titel, x label, y label, stick, ytick
fig=plt.figure()
ax= fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000). cumsum(), label='Random cumsum' ,marker='o') #o for round, s for squere , d for dimond
ax.set_xticks([0,250,500,750,1000]) #customize the ticks on x coordinate
ax.set_xticklabels(['one','two', 'three', 'four', 'five'], rotation=30, fontsize='small') #use labels for ticks
ax.set_title('the cumsum plot') #create a title
ax.set_xlabel('stages') #create a lebel for the coordinate
ax.set_ylabel('values')
```

```python
#one more exmple
fig= plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'r--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'b', label='three')
ax.legend(loc='best')
ax.text(10,20, 'Hello world', family='monospace', fontsize=10)
```

```python
#in this example, we practice bar chart
fig, axes = plt.subplots(2,1)
data=pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0],color='k', alpha=1.0) #alpha is for transparncy, 1.0 means not transparent
data.plot.barh(ax=axes[1],color ='r',alpha=0.2)
```
