```python
import pandas as pd
```

```python
airbnb =pd.read_csv('Airbnb_Texas_Rentals.csv', index_col=0)
```

```python
airbnb['average_rate_per_night']=airbnb['average_rate_per_night'].str.replace('$','')
```

```python
airbnb['average_rate_per_night']= airbnb['average_rate_per_night'].astype(float)
```

```python
pd.isnull(airbnb['bedrooms_count']).any()
```

```python
airbnb.dropna(subset=['bedrooms_count'], inplace=True)
```

```python
#next, we want to see the average of rate for each bedroom type.
```

```python
grouped_bedrooms =airbnb['average_rate_per_night'].groupby(airbnb['bedrooms_count'])
```

```python
grouped_bedrooms.mean()
```

```python
#next we want to know 
grouped_bedrooms.size()
```

```python
grouped_bedrooms.max()
```

```python
grouped_bedrooms.min()
```

```python
grouped_bedrooms.describe()
```

```python
#we want to create a pivot tbale betwwn bedrooms count and city
```

```python
airbnb.pivot_table(['average_rate_per_night'], index='bedrooms_count', columns='city')
```
