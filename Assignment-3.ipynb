```python
import pandas as pd
```

```python
#Question 1
airbnb =pd.read_csv('Airbnb_Texas_Rentals.csv', index_col=0)
airbnb.head()
```

```python
#Question 2
avg_rate_missed = airbnb['average_rate_per_night'].isna()
avg_rate_missed.any()
```

```python
#Qurstion 3
avg_nan_value = airbnb['average_rate_per_night'].isna()
airbnb[avg_nan_value]
```

```python
#Question 4
count_missed = airbnb['bedrooms_count'].isna()
airbnb[count_missed]
```

```python
#Question 5
airbnb.dropna(subset = ['average_rate_per_night','bedrooms_count'], inplace=True)
```

```python
#question 6
airbnb['city'].fillna('Not Available', inplace = True)
airbnb['date_of_listing'].fillna('Not Available', inplace = True)
airbnb['description'].fillna('Not Available', inplace = True)
airbnb['title'].fillna('Not Available', inplace = True)
airbnb['url'].fillna('Not Available', inplace = True)
```

```python
airbnb.isna().any()
```

```python
#Question 7
airbnb['average_rate_per_night'] = airbnb['average_rate_per_night'].str.replace('$','').astype(int)
airbnb.head()
```

```python
(airbnb['average_rate_per_night'] < 0).any()
all_mean = airbnb['average_rate_per_night'].mean()
all_mean
```

```python
#question 9
type_bdrooms =airbnb.bedrooms_count.unique()
type_bdrooms
```

```python
#Question 10
stud_bdrooms= (airbnb['bedrooms_count'] =='Studio')
stud_bdrooms.any()
```

```python
stud_bdrooms2 =airbnb[stud_bdrooms]
stud_bdrooms_avg =stud_bdrooms2['average_rate_per_night'].mean()
stud_bdrooms_avg
```

```python
#question 11
type_bdrooms[type_bdrooms == 'Studio'] ='0'
mx_bdrooms=type_bdrooms.max().astype(int).astype(str)
mx_bdrooms
```

```python
mx_bdroom_avg =airbnb[airbnb['bedrooms_count'] == mx_bdrooms].average_rate_per_night.astype(int).mean()
mx_bdroom_avg
```
