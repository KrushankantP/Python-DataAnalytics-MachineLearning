## Q3.
Show a sum of profit of each product sub-category sold in Texas and also a bar chart as follows. 

```python
Tx=ss[ss['State']=='Texas']
profit_Tx=Tx.groupby(Tx['Product Sub-Category'])['Profit'].sum()
```

```python
profit_Tx
```

```python
ax=profit_Tx.plot.bar()
ax.set_title('Texas Sales Profit Analysis')
ax.set_xlabel('Product Sub-Category')
ax.set_ylabel('Profit')
```

## Q4. 
Display the data type of the values in the column of ‘Order Date’. If there is a need to change the column into a date, use an appropriate function to do it. If there is no need to do it, do not do it. Create a pivot table regrading the total sales of each state in each year. Then, draw a line chart to show the results.

```python
#Q4:
sale_od=ss.pivot_table('Sales',index=ss['Order Date'].dt.year,columns='State',aggfunc='sum')
```

```python
sale_od
```

```python
ax=sale_od.plot(figsize=(10,5),marker='s')
ax.set_title('Total sales of each state from 2009 to 2012')
ax.set_xlabel('Year')
ax.set_ylabel('States')
```

## Q5.

## Q7. 
Display the top 10 customers who made largest sales in 2011.  

```python
#Q-7
top10=ss[ss['Order Date'].dt.year==2011]
top10_cust=top10.groupby(top10['Customer Name'])['Sales'].sum()
top10_cust.sort_values(ascending=False).head(10)
```

## Q10.
Create a figure with two subplots in which a subplot shows a bar chart of the average discount rates by product categories and another subplot shows a bar chart of the average discount rates by customer segments. 

```python
#Q-10
Avg_dis1=ss.pivot_table('Discount',index=ss['Order Date'].dt.year,columns='Product Category',aggfunc='mean')
Avg_dis2=ss.pivot_table('Discount',index=ss['Order Date'].dt.year,columns='Customer Segment',aggfunc='mean')
```

```python
fig, ax=plt.subplots(1,2,figsize=(15,5))
Avg_dis1.plot.bar(ax=ax[0])
Avg_dis2.plot.bar(ax=ax[1])

ax[0].set_title('Discount Rates by Product Category')
ax[0].set_xlabel('year')
ax[0].set_ylabel('Discount Rate')

ax[1].set_title('Discount Rates by customer Segment')
ax[1].set_xlabel('year')
ax[1].set_ylabel('Discount Rate')
```
