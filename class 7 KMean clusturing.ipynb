```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nba =pd.read_excel('NBA_Players_2017.xlsx', index_col =0)
```

```python
nba
```

```python
#now we want to create a dataframe named PG_c that containts only PG and  C player
PG_C=nba[(nba['Pos'] == 'PG') |(nba['Pos']== 'C')]
PG_C.head()
```

```python

PG_C=PG_C[['Player','Pos', 'BPG', 'APG']]
PG_C =PG_C[(PG_C['BPG']>0.0)|(PG_C['APG']>0.0)]
```

```python
PG_C
```

```python
#WE WANT TO USE SEABORN TO DRAW A SCATTER PLAT
import seaborn as sns
scatterData =PG_C[['APG','BPG']]
sns.scatterplot(x='APG', y='BPG', data=scatterData)
```

```python
# use scikitlearn to perform K-mean clustering
from sklearn.cluster import KMeans
clut=KMeans(n_clusters =2, random_state =0).fit(PG_C[['APG','BPG']])
```

```python
clut.cluster_centers_
```

```python
clut.labels_
```

```python
labels =['C' if i==0 else 'PG' for i in clut.labels_]
labels
```

```python
#Add labels to the dataframe
PG_C['Label'] =labels
PG_C
```

```python
#we want to calculate the accuracy rate
PG_C.shape
```

```python
sum(PG_C['Label']==PG_C['Pos'])/PG_C.shape[0]
```

```python
fig =plt.figure(figsize=(6,6))
ax=fig.add_subplot(1,1,1, facecolor='w')
C=PG_C[PG_C['Pos']=='C']
ax.scatter(C['APG'],C['BPG'],color='b',s=150)
PG = PG_C[PG_C['Pos']=='PG']
ax.scatter(PG['APG'],PG['BPG'],color='r',s=150)
plt.xlabel('Assists per game')
plt.ylabel('Anaythibg')
plt.title('Antithing')
plt
```

```python
x=PG_C[['APG','BPG']]
x.head()
```

```python
y=PG_C['Pos']
y.head()
```

```python
#when we use logistic regression, x should be array like structure
x=x.values
x
```

```python
y=y.values
y
```

```python
#next create split sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state =42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train, y_train)
```

```python
predictions =knn.predict(x_test)
```

```python
sum(predictions ==y_test)/len(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
reg.fit(x_train, y_train)
```

```python
pred=reg.predict(x_test)
pred
```

```python
y_test
```

```python
sum(pred==y_test)/len(pred)
```
